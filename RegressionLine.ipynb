```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.plot.scatter(x='sepal_length', y='petal_length', title='Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
tips.plot.scatter(x='total_bill', y='tip', title='Iris Dataset')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
sns.regplot(x="total_bill", y="tip", data=tips)
pearson_r = tips['total_bill'].corr(tips['tip'])
pearson_r
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=iris)
sns.regplot(x="sepal_length", y="petal_length", data=iris)
```

```python
pearson_r = iris['sepal_length'].corr(iris['petal_length'])
pearson_r
```
