```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.plot.scatter(x='sepal_length', y='petal_length', title='Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
import numpy as np
```

```python
# create a pandas DataFrame with two columns
df = pd.DataFrame(columns=['weight', 'chocolate consumption'])

# fill the columns with 50 random values
df['weight'] = np.random.randint(50, 100, 50)
df['chocolate consumption'] = np.random.randint(0, 10, 50)
```

```python
df
```

```python
sns.scatterplot(x="weight", y="chocolate consumption", data=df)
sns.regplot(x="weight", y="chocolate consumption", data=df)
pearson_r = df['weight'].corr(df['chocolate consumption'])
pearson_r
```

# note, the above is a random dataset

```python
# create a pandas DataFrame with two columns
df = pd.DataFrame(columns=['weight', 'chocolate consumption'])

# fill the columns with 50 random values
df['weight'] = np.random.normal(loc=70, scale=10, size=50)
df['chocolate consumption'] = df['weight']*0.05 + np.random.normal(loc=2, scale=1, size=50)
```

```python
df
```

```python
df.plot.scatter(x='weight', y='chocolate consumption', title='bodyweight/chocolate relationship')
plt.xlabel('weight')
plt.ylabel('chocolate consumption')
plt.show()
```

```python
# create a pandas DataFrame with two columns
df = pd.DataFrame(columns=['weight', 'chocolate consumption'])

# fill the columns with 50 random values
df['weight'] = np.linspace(60, 100, 50)
df['chocolate consumption'] = df['weight']*0.06 + np.random.normal(loc=0, scale=1.5, size=50)
df.plot.scatter(x='weight', y='chocolate consumption', title='bodyweight/chocolate relationship')
plt.xlabel('weight')
plt.ylabel('chocolate consumption')
plt.show()
```
