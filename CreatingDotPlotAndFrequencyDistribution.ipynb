```python
import pandas as pd
```

```python
data = {'school': ['1', '2', '3', '4', '5','6','7','8'],
        'average': [7.4, 7.9, 4.1, 8.1, 6.2,7.1,7.4,6.7]}

df = pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a dot plot
df.plot(x='average', y='school', kind='scatter')

# Show the plot
plt.show()
```

# frequency dis

```python
# Create a frequency distribution table
freq_table = df['average'].value_counts().sort_index()
print(freq_table)

# Create a histogram
df.hist(column='average', bins=range(1,9), rwidth=0.8, color='#86bf91')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of A')

# Show the plot
plt.show()
```
