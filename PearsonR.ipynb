```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.plot.scatter(x='sepal_length', y='petal_length', title='Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
# Compute the Pearson's correlation coefficient between sepal_length and petal_length
pearson_r = iris['sepal_length'].corr(iris['petal_length'])
```
